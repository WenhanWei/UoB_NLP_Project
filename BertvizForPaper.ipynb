```python
from transformers import AutoTokenizer, AutoModel
```

```python
tweet = "Oklahomas first confirmed case of coronavirus is in Tulsa County  SmartNews"
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-large-cased")
model = AutoModel.from_pretrained("bert-large-cased", output_attentions=True)
inputs = tokenizer.encode(tweet, return_tensors='pt', add_special_tokens=False)
outputs = model(inputs)
attention = outputs[-1]  # Output includes attention weights when output_attentions=True
tokens = tokenizer.convert_ids_to_tokens(inputs[0]) 
```

```python
from bertviz import head_view
head_view(attention, tokens)
```

```python
from bertviz import model_view
model_view(attention, tokens, display_mode="light")
```
